```python
import pandas as pd
df = pd.read_csv("/content/AB_NYC_2019.csv")
df.head()
```

#Detect outliers using percentile

```python
df.price.describe()
```

```python
# Calculate the 1st and 99th percentiles of the price
lower_limit = df['price'].quantile(0.01)
upper_limit = df['price'].quantile(0.99)

# Filter out the outliers
filtered_data = df[(df['price'] > lower_limit) & (df['price'] < upper_limit)]

# Check the shape of the original and new dataframes
print("Original DataFrame shape:", df.shape)
print("New DataFrame shape:", filtered_data.shape)

# Optionally, you can view the first few rows of the new dataframe
filtered_data.head()
```
